```python
# get PAS in bed, intersect with DB

# step1, get uniq pas number and freq
import re,time
# ## uniq
# $ cat 225.bed |awk '{if($6=="+") print $1":"$3":"$6; else print $1":"$2":"$6}'|sort |uniq -c >225.uniq.pas
# shell too slow, using Python
fr=open(bedTotal,'r')
i=0
start=time.time();
DB={}
while True:
    lineR=fr.readline(); #225706775
    if not lineR:
        break;
    i+=1;
    if i>1000000:
        #break;
        pass;
    if i%5706775==0:
        print(i,' line procced.', round(time.time()-start,2),'seconds elapsed')
    line=lineR.strip()
    arr=re.split("\t",line)
    #
    k=arr[0]+":"+arr[5]
    if arr[5]=='+':
        pos=arr[2]
    elif arr[5]=="-":
        pos=arr[1]
    #
    if k not in DB:
        DB[k]={};
    if pos not in DB[k]:
        DB[k][pos]=0
    DB[k][pos]+=1;
#get PAS file
#

fr.close()
timeString=time.strftime("%Y%m%d-%H%M%S", time.localtime()) 
print("==end==",i, round(time.time()-start,2),'seconds elapsed', timeString)
print(len(DB), DB.keys())
#last for 14min
```

```python
# step2ï¼Œsave uniq PAS sites
# bed file column 5th:count, 2nd-3rd: the same position.
##chr1	564599	564599	chr1:564599:+	3.33339746426	+
pas_PY="../bed/225.pas_PY.bed"
fw=open(pas_PY,'w')
i=0
for chrStrand in DB:
    arr=re.split(':',chrStrand);
    #print(arr_chrStrand, len(DB[chrStrand])) #['chr10', '-'] 61299
    for pos in DB[chrStrand]:
        i+=1;
        if i>10:
            #break;
            pass;
        count=DB[chrStrand][pos];
        #print(item, DB[chrStrand][pos]) #101688 1
        arr2=[arr[0], pos,pos, arr[0]+":"+pos+':'+arr[1], str(count), arr[1]] #['chr10', '101688', '101688', 'chr10:101688:-', '1', '-']
        #print(arr2)
        fw.write('\t'.join(arr2) +'\n');
fw.close()
print('==End==',i)
```

```python
# step3 load polyA_DB3, and uniq PAS bed saved just now.
import re
def getDBtoMemo(filePath):
    start=time.time();
    fr=open(filePath, 'r')
    i=0
    mydict={}
    for lineR in fr.readlines():
        i+=1
        if i>100000:
            #break;
            pass;
        line=lineR.strip();
        arr=re.split('\t',line);
        #print(arr)
        k=arr[0]+':'+arr[-1]
        if k not in mydict:
            mydict[k]=[];
        mydict[k].append(int(arr[1])) # number type

    print('i=',i,len(mydict),mydict.keys() );
    timeString=time.strftime("%Y%m%d-%H%M%S", time.localtime()) 
    print( round(time.time()-start, 2) ,' seconds elapsed. ', timeString)
    fr.close()
    return mydict
#
polyA_DB3_dict=getDBtoMemo(polyA_DB3)
# pas_PY="../bed/225.pas_PY.bed"
print('-'*50)
pas_PY_dict=getDBtoMemo(pas_PY)
print('-'*50)
```

```python
# [more rapid method! 2000 fold faster]
# step4 get intersect of Uniq PAS site, and polyA_DB3, parameter: distance

import time
# input distance, give number of intersect(PAS, DB), count if within a given distance.
def getUnionNumberWithDB(distance):
    siteNumber=0
    for chrStrand in pas_PY_dict:
        if chrStrand not in polyA_DB3_dict:
            continue;
        #
        #print(chrStrand, 'begin','='*20)
        # sort by strand and chr
        pas_PY_dict[chrStrand].sort();
        polyA_DB3_dict[chrStrand].sort();
        # define about the DB chr:strand array
        maxPos=len(polyA_DB3_dict[chrStrand])-1;# max coord in the current chr
        pointer=0;# pointer to DB, the index of arrray
        #
        for pos in pas_PY_dict[chrStrand]:
            while polyA_DB3_dict[chrStrand][pointer]< pos-distance: 
                # if pointer to DB < current lower limit, move to next
                pointer+=1;
                #fix: must NOT bigger than max coordination
                if pointer>maxPos: 
                    break;

            #fix 
            if pointer>maxPos:
                break;
            if pos+distance >= polyA_DB3_dict[chrStrand][pointer]:
                siteNumber+=1; # if find, +1; else do nothing.
        #print(chrStrand, 'end','; siteNumber=',siteNumber)
        #print( round(time.time()-start, 2) ,' seconds elapsed. ', '-'*20)
    #
    return siteNumber;
    #print('='*50,'distance=',distance, '; siteNumber =',siteNumber)
#
start=time.time();

#test
distance=0
print(distance, getUnionNumberWithDB(distance) );

#
timeString=time.strftime("%Y%m%d-%H%M%S", time.localtime()) 
print( round(time.time()-start, 2) ,' seconds elapsed. ', timeString)
#1.14s; siteNumber = 416347
```

```python
#step 5: batch mode. then using ggplot2 plot.
with open('UnionNumberOfPASWithDB_by_distance.txt','w') as f:
    for distance in range(0,101, 5):
        rs=getUnionNumberWithDB(distance);
        print(distance, rs );
        f.write(str(distance)+'\t'+str(rs)+'\n' )
# distance apa_number
```
