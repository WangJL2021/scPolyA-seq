# step1 get bed from pas site

```python
#1. get XX nt sequence up and down, and find motif
import re,time
basePath="/home/wangjl/data/apa/191111Figure/f2/validation/"

def getBed(fname, up=0, down=0, keyword="", region=""): 
    fr=open(fname,"r")
    if keyword=="":
        keyword=region;
    
    #chr10	855482	855556	chr10:855482:-	36	-
    #chr10	157706	157706	chr10:157706:-	5	-
    fw=open(basePath+'pas_bed/up'+str(up)+'_down'+str(down)+"_"+keyword+".bed", 'w')
    #错误的行
    fwE=open(basePath+'pas_bed/up'+str(up)+'_down'+str(down)+"_E_"+keyword+".bed", 'w')
    #
    begin=time.time()
    lines=fr.readlines()
    i=0
    new_i=0
    for lineR in lines:
        i+=1
        if i==1: #no header line
            continue;
        if i>10:
            #break;
            pass;
        
        line=lineR.strip()
        #hint 10 times
        if i % round(len(lines)/10)==0:
            print(i, round(time.time()-begin,2),"seconds elapsed.")
        #chr|strand as key
        pas=re.split("\s",line)
        #print(pas)
        #['chr10:320403:-', 'chr10', '320403', '320403', 'chr10:320403:-', '10945', '-', 'UTR3', 'DIP2C-201', 'TGAAAATAGCAGTTTCTTAAT', '0']
        
        
        # filter by region
        if region=="":
            pass;
        elif region!=pas[7]:
            continue;
        
        arr=re.split(":",pas[0])
        
        # discard MT RNA
        if arr[0]=="chrM":
            #print(line);
            fwE.write(lineR);
            continue;
        
        ################################################ important! for test only.
        #if arr[2]=="-": # get chain specific seq
        #    continue;
            
        if arr[2]=="+":
            pos=int(arr[1]) ##+ should minus 1! cause bam to bed plus 1
            start=pos-up-1;
            end=pos+down+1-1;
        else:
            pos=int(arr[1]) 
            start=pos-down; 
            end=pos+up+1
        ##fix
        if start<=0:
            print(line);
            fwE.write(lineR);
            continue;
        ##
        new_i+=1
        arr2=[arr[0], str(start), str(end), pas[0], pas[5], arr[2]]
        bedStr='\t'.join(arr2)
        #print( arr, bedStr ) 
        #['chr10', '855480', '855482', 'chr10:855482:-', '36', '-'] chr10	855480	855482	chr10:855482:-	36	-
        
        #freq=int(arr[4])  fw.write(  ( bedStr +'\n')*freq )
        fw.write( bedStr +'\n' )
    #
    fr.close(); fw.close(); fwE.close();
    print("==end==",pasFileName, i, new_i, region)

#test
pasFileName="/data/jinwf/wangjl/apa/190705PAS/bed/pasPostions_Location_transcriptName-noChrM_noInnerPrime_PY-filterCountCell.bed"

#test
getBed(pasFileName, 100, 100, keyword='all')
#getBed(pasFileName, 100, 100, keyword='minus')
```

```python
#2.bed 2 seq
# version 2.0 auto filename
import subprocess

def getSeqFromBed(up=0, down=0, keyword=''):
    fastaPath=basePath+"pas_bed/"

    nlen=str( up+down )
    print(keyword,up, down)
    cmd1="bedtools getfasta -fi /home/wangjl/data/ref/hg19/hg19.fa -bed "+fastaPath+ \
        "up"+str(up)+"_down"+str(down)+'_'+keyword+".bed -s -fo "+fastaPath+nlen+'_'+keyword+".fasta"
    cmd2='grep -v ">" '+fastaPath+nlen+'_'+keyword+'.fasta > '+fastaPath+nlen+'_'+keyword+'.fa'
    print(cmd1)
    print(cmd2)

    (status1, output1)=subprocess.getstatusoutput(cmd1)
    (status2, output2)=subprocess.getstatusoutput(cmd2)

    print(status1, output1)
    print(status2, output2)

    import time
    print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
#
getSeqFromBed(100,100,'all')
getSeqFromBed(60,0,'all')

for region in regions:
    print('='*10, region)
    getSeqFromBed(100, 100, region)
    getSeqFromBed(60, 0, region)

#test
#getSeqFromBed(100,100,'plus')
#getSeqFromBed(100,100,'minus')

print('==end==')
```

# step3 base freq

```python
# get base freq from seq file
# v2.0
import re,time

def getFreq_fromSeq(fname):
    A=[0]*201
    T=[0]*201
    G=[0]*201
    C=[0]*201
    N=[0]*201

    fr=open(basePath+'/pas_bed/'+fname+'.fa','r')
    #line="aACTATTGGTAATTTAAGTTTCCTTTTGTGTCATTCAGTGTAAAACTGTCTAATTTGAAAAAAAATGTAGGTTATGAAAATAAAGATTTAGGCACTGTTCAATTTTTCTTTTGattttttttaaattaaaatttCTTCAATAATGCATCTTTTCTAAACATAGTTTTTTGTCTGTTCCCATAAGTGCATGAGAAAGAAACA"
    #print(line)
    lines=fr.readlines();
    for lineR in lines:
        line=lineR.strip().upper();
        #base freq
        for i in range(len(line)):
            #print(i, line[i])
            if line[i]=="A":
                A[i]+=1
            elif line[i]=="T":
                T[i]+=1
            elif line[i]=="G":
                G[i]+=1
            elif line[i]=="C":
                C[i]+=1
            else:
                N[i]+=1

    #
    print("base sum=", sum(A)+sum(T)+sum(G)+sum(C)+sum(N), ' should be at least:', len(lines)*len(line))
    print("base sum:(A/T/G/C/)", sum(A), sum(T),sum(G),sum(C),sum(N) )
    print(sum(A), "A=",A)
    #print(sum(T), "T=",T)
    #print(sum(G), "G=",G)
    #print(sum(C), "C=",C)
    #print(sum(N), "N=",N) #N no base

    fr.close()

    #############
    #freq ratio at each site
    a=["0"]*201
    t=["0"]*201
    g=["0"]*201
    c=["0"]*201

    for i in range(len(A)):
        total=A[i]+T[i]+G[i]+C[i]
        if total>0:
            a[i]=str(A[i]/total)
            t[i]=str(T[i]/total)
            g[i]=str(G[i]/total)
            c[i]=str(C[i]/total)
    #print("a=",a, '\n')

    # save
    fw=open(basePath+'/pas_bed/'+fname+'.fa.freq','w')
    fw.write('A\t'+"\t".join(a)+"\n")
    fw.write('U\t'+"\t".join(t)+"\n")
    fw.write('G\t'+"\t".join(g)+"\n")
    fw.write('C\t'+"\t".join(c)+"\n")
    fw.close()
    #
    print('=Freq end:',fname)
#
#test
getFreq_fromSeq('200_all')
for region in regions:
    print('='*10, region)
    getFreq_fromSeq('200_'+region)
print('==end==')
```
