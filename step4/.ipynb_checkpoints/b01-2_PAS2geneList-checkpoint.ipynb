```python
# [Run once]

#
fr=open("/home/wangjl/data/ref/hg19/gencode.v30lift37.annotation.GENE.gtf",'r')

i=0
g=0 #gene number
refer={}
for lineR in fr.readlines():
    if lineR.startswith("#"):
        continue;
    i+=1;
    if i>5:
        pass;
        #break;
    if i%50000==0:
        print("i=%d, g=%d, g/i=%f" % (i,g, g/i*100))

    line=lineR.strip()
    arr=re.split('\t',line)
    
    #get gene symbol from last col
    arr2=re.split("gene_name ", arr[-1])
    arr3=re.split("; ",arr2[1])
    #print(arr[0], arr[6],  arr[3], arr[4],  arr3[0]) #chr1 11869 14409 + "DDX11L1"
    arr3[0]=arr3[0].strip('"')
    
    # make refer
    k=arr[0]+arr[6]
    if k not in refer:
        refer[k]=[]
    #
    refer[k].append([int(arr[3]), int(arr[4]), arr3[0] ])
        
    #print(arr)
    #['chr1', 'HAVANA', 'gene', '11869', '14409', '.', '+', '.', 
    #'gene_id "ENSG00000223972.5_2"; gene_type "transcribed_unprocessed_pseudogene"; gene_name "DDX11L1"; level 2; 
    #havana_gene "OTTHUMG00000000961.2_2"; remap_status "full_contig"; remap_num_mappings 1; remap_target_status "overlap";']

fr.close()
print(refer.keys()) 
#{'chr1+': [[11869, 14409, '"DDX11L1"'], [29554, 31109, '"MIR1302-2HG"'], [52473, 53312, '"OR4G4P"']], 
# 'chr1-': [[14404, 29570, '"WASH7P"'], [34554, 36081, '"FAM138A"']]}
print("==end==")
```

```python
## assign gene symbol to PAS site
def pas2gene(site):
    #site=pas[0]
    if site[0]+site[2] not in refer:
        return ".";
    referSite=refer[site[0]+site[2]]
    for arr in referSite:
        if site[1]>arr[0] and site[1]<arr[1]:
            return arr[2]
    return "."
print( pas2gene(pas[0]) )



# simple get
def get22():
    genes=set()
    i=0
    for p in pas:
        gene=pas2gene(p);
        if gene!="":
            i+=1;
            if gene in genes:
                print(gene)
            genes.add(gene)
            #print(p, gene)
    #
    print('输入：', len(pas))
    print(i,len(genes), genes)
#
```
