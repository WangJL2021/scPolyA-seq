```python
# Load small bed of each cell, using PAS site in json, get freq of each site in this cell
# steps：
#1. load small bed，chr1:+  {1:count1,  29:count2, }
#2. using apa refer index JSON combine counts in a cluster, save results to peak in index, output 2 col like chr1:100:+  123 
#   not find in small bed, note as 0
#3. do in parallel

import os
os.chdir("/data/jinwf/wangjl/apa/190705PAS/pdf")
os.getcwd()
```

```python
# step1: load small bed of each cell, to a dict bedDB
# key is chr1:+ 
# value is an array of dict {1:count1, 29:count2, ...}
import re,time

def readBed(fid):
    start=time.time();
    fr=open('/data/jinwf/wangjl/apa/190705PAS/bed/'+fid+'.bed','r');
    i=0
    bedDB={}
    for lineR in fr.readlines():
        i+=1
        if i>10:
            #break;
            pass;
        line=lineR.strip();
        arr=re.split('\t', line);
        #print(arr) #['chr10', '300469', '300566', 'ST-E00167:377:H52JTALXX:8:2202:22039:22194', '255', '+']
        #            ['chr10', '856735', '856853', 'ST-E00167:377:H52JTALXX:8:2216:8410:1608', '255', '-']
        #create (key)chrStrand in bedDB
        chrStrand=arr[0]+":"+arr[-1]
        if chrStrand not in bedDB:
            bedDB[chrStrand]={}
        #
        if arr[-1]=="+":
            pos =arr[2];
        else:
            pos=arr[1];
        #whether pos in chr1:+:{ this }
        if pos in bedDB[chrStrand]:
            bedDB[chrStrand][pos]+=1;
        else:
            bedDB[chrStrand][pos]=1
        #
    fr.close();
    timeString=time.strftime("%Y%m%d-%H%M%S", time.localtime());
    print("readBed(): fid=",fid, '; i=',i,'; ','; 耗时',round(time.time()-start,2),'s; ', timeString, sep='')
    # readBed(): fid=c01_ROW07; i=1223251; ; 耗时2.57s; 20190928-222958
    return (i,bedDB); #返回总行数，和该文件的字典形式
#test
fid="c01_ROW07"
i,bedDB1=readBed(fid)
```

```python
# load referDB JSON
import json
with open("/data/jinwf/wangjl/apa/190705PAS/bed/pasPostions_siteAndIncludes_PY.json",'r') as load_f:
    pasPostions = json.load(load_f)
    print(type(pasPostions))
#<class 'dict'> 
```

```python
#step2: Based on small bedDB, refer to referDB, get counts for each row in referDB
import re;
def savePASbyID(fid,lineNo,bedDB1):
    start= time.time()
    i=0
    j=0 #counter
    fw=open('/data/jinwf/wangjl/apa/190705PAS/bed/freq2/'+fid+'.freq','w')
    for chrStrand in pasPostions:
        chr1,strand=re.split(':',chrStrand);
        for site in pasPostions[chrStrand]:
            i+=1;
            if i>1000:
                #break;
                pass;
            key=chr1+":"+site+":"+strand;
            count=0;
            #print('pasPostions[chrStrand][site]=',pasPostions[chrStrand][site])

            for pos in pasPostions[chrStrand][site]:
                pos=str(pos);
                if chrStrand in bedDB1:
                    if pos in bedDB1[chrStrand]:
                        count+=bedDB1[chrStrand][pos]
                        #bedDB1[chrStrand][pos]=0 #破坏性的验证过，都是等于1的位点
            #
            fw.write('\t'.join([key, str(count)]) + '\n')
            j+=count;
    fw.close();

    timeString=time.strftime("%Y%m%d-%H%M%S", time.localtime());
    print("savePASbyID(): fid=",fid,'; PAS site number i=',i, '; cover counts j=',j,'(',round(j/lineNo*100,3),'%); Consume time=',round(time.time()-start,2),'s; ', timeString, sep='')
    #savePASbyID(): fid=c01_ROW07; i=742288; j=1220056; ; 耗时1.27s; 20190929-104357
    #j smaller, as filter out some read=1 sites
savePASbyID(fid,i, bedDB1)
```

```python
# step3: Parallel
import sys
sys.path.append("/home/wangjl/test/")
from ParallelMe import ParallelMe

# single cmd, depend on id
def doCMD(fid):
    lineNo,bedDB1=readBed(fid)
    savePASbyID(fid, lineNo, bedDB1)
    return 0;
##test
#doCMD('c01_ROW07')

#id list file
cIDs="/data/jinwf/wangjl/apa/190705PAS/225.cellID"

#parallel
myTasks=ParallelMe(doCMD, cIDs, core=100);
myTasks.run()

# 225 items processed in 16.04 seconds
print('===end===')
```
